```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

# ==========================================
# TẠO DỮ LIỆU
# ==========================================

means = [[2, 2], [9, 2], [4, 9]]

cov = [[2, 0],
       [0, 2]]

n_samples = 500
n_cluster = 3

X0 = np.random.multivariate_normal(
    means[0],
    cov,
    n_samples
)

X1 = np.random.multivariate_normal(
    means[1],
    cov,
    n_samples
)

X2 = np.random.multivariate_normal(
    means[2],
    cov,
    n_samples
)

X = np.concatenate((X0, X1, X2), axis=0)

# ==========================================
# HIỂN THỊ DỮ LIỆU
# ==========================================

plt.xlabel('x')
plt.ylabel('y')

plt.plot(
    X[:,0],
    X[:,1],
    'bo',
    markersize=4
)

plt.title("Du lieu ban dau")

plt.show()

# ==========================================
# KHỞI TẠO TÂM CỤM
# ==========================================

def kmeans_init_centers(X, n_cluster):

    return X[
        np.random.choice(
            X.shape[0],
            n_cluster,
            replace=False
        )
    ]

# ==========================================
# GÁN NHÃN
# ==========================================

def kmeans_predict_labels(X, centers):

    D = cdist(X, centers)

    return np.argmin(D, axis=1)

# ==========================================
# CẬP NHẬT TÂM CỤM
# ==========================================

def kmeans_update_centers(
        X,
        labels,
        n_cluster):

    centers = np.zeros(
        (n_cluster, X.shape[1])
    )

    for k in range(n_cluster):

        Xk = X[labels == k, :]

        centers[k,:] = np.mean(
            Xk,
            axis=0
        )

    return centers

# ==========================================
# KIỂM TRA HỘI TỤ
# ==========================================

def kmeans_has_converged(
        centers,
        new_centers):

    return (
        set([tuple(a) for a in centers]) ==
        set([tuple(a) for a in new_centers])
    )

# ==========================================
# VẼ KẾT QUẢ
# ==========================================

def kmeans_visualize(
        X,
        centers,
        labels,
        n_cluster,
        title):

    plt.xlabel('x')
    plt.ylabel('y')

    plt.title(title)

    colors = ['b', 'g', 'r', 'c', 'm', 'y']

    for i in range(n_cluster):

        data = X[labels == i]

        plt.plot(
            data[:,0],
            data[:,1],
            colors[i] + '^',
            markersize=4,
            label='cluster_' + str(i)
        )

        plt.plot(
            centers[i][0],
            centers[i][1],
            colors[i+3] + 'o',
            markersize=10,
            label='center_' + str(i)
        )

    plt.legend()

    plt.show()

# ==========================================
# THUẬT TOÁN KMEANS
# ==========================================

def kmeans(
        init_centers,
        init_labels,
        X,
        n_cluster):

    centers = init_centers

    labels = init_labels

    times = 0

    while True:

        labels = kmeans_predict_labels(
            X,
            centers
        )

        kmeans_visualize(
            X,
            centers,
            labels,
            n_cluster,
            'Gan nhan lan ' + str(times + 1)
        )

        new_centers = kmeans_update_centers(
            X,
            labels,
            n_cluster
        )

        if kmeans_has_converged(
                centers,
                new_centers):
            break

        centers = new_centers

        kmeans_visualize(
            X,
            centers,
            labels,
            n_cluster,
            'Cap nhat tam cum lan ' + str(times + 1)
        )

        times += 1

    return (
        centers,
        labels,
        times
    )

# ==========================================
# CHẠY CHƯƠNG TRÌNH
# ==========================================

init_centers = kmeans_init_centers(
    X,
    n_cluster
)

print("Tam cum khoi tao:")

print(init_centers)

init_labels = np.zeros(X.shape[0])

kmeans_visualize(
    X,
    init_centers,
    init_labels,
    n_cluster,
    'Khoi tao ban dau'
)

centers, labels, times = kmeans(
    init_centers,
    init_labels,
    X,
    n_cluster
)

print("Kmeans hoi tu sau",
      times,
      "lan lap")

print("Tam cum cuoi cung:")

print(centers)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

# ==========================================
# TẠO DỮ LIỆU
# ==========================================

X, y = make_blobs(
    n_samples=100,
    n_features=2,
    centers=4,
    cluster_std=1,
    random_state=4
)

# ==========================================
# HIỂN THỊ DỮ LIỆU
# ==========================================

plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    marker='o',
    s=50
)

plt.title("Du lieu ban dau")

plt.show()

# ==========================================
# CHIA TRAIN / TEST
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state=0
)

# ==========================================
# KNN VỚI k = 5
# ==========================================

knn5 = KNeighborsClassifier(
    n_neighbors=5
)

knn5.fit(
    X_train,
    y_train
)

y_pred_5 = knn5.predict(X_test)

plt.figure(figsize=(8,6))

plt.scatter(
    X_test[:,0],
    X_test[:,1],
    c=y_pred_5,
    marker='o',
    s=50
)

plt.title("KNN voi k = 5")

plt.show()

# ==========================================
# KNN VỚI k = 1
# ==========================================

knn1 = KNeighborsClassifier(
    n_neighbors=1
)

knn1.fit(
    X_train,
    y_train
)

y_pred_1 = knn1.predict(X_test)

plt.figure(figsize=(8,6))

plt.scatter(
    X_test[:,0],
    X_test[:,1],
    c=y_pred_1,
    marker='o',
    s=50
)

plt.title("KNN voi k = 1")

plt.show()

# ==========================================
# TÌM k TỐT NHẤT
# ==========================================

knn_grid = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid={
        'n_neighbors': np.arange(1,10)
    },
    cv=5
)

knn_grid.fit(X, y)

print("k tot nhat:")

print(knn_grid.best_params_)

# ==========================================
# CÀI ĐẶT KNN THỦ CÔNG
# ==========================================

def KNN(
        X_train,
        X_test,
        y_train,
        k):

    num_test = X_test.shape[0]

    num_train = X_train.shape[0]

    y_pred = np.zeros(
        (num_test, num_train)
    )

    for i in range(num_test):

        for j in range(num_train):

            y_pred[i,j] = np.sqrt(
                np.sum(
                    np.power(
                        X_test[i,:] -
                        X_train[j,:],
                        2
                    )
                )
            )

    results = []

    for i in range(len(y_pred)):

        zipped = zip(
            y_pred[i,:],
            y_train
        )

        res = sorted(
            zipped,
            key=lambda x:x[0]
        )

        results_topk = res[:k]

        classes = {}

        for _, j in results_topk:

            j = int(j)

            if j not in classes:
                classes[j] = 1
            else:
                classes[j] += 1

        results.append(
            max(
                classes,
                key=classes.get
            )
        )

    return np.array(results)

# ==========================================
# TEST KNN
# ==========================================

X_test_new = np.array([
    (1,3)
])

results = KNN(
    X,
    X_test_new,
    y,
    3
)

print("Nhan du doan:")

print(results)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# ==========================================
# DỮ LIỆU KHÁCH HÀNG
# ==========================================

X = np.array([
    [15,39],
    [16,81],
    [17,6],
    [18,77],
    [19,40],
    [20,76],
    [21,6],
    [22,94]
])

# ==========================================
# KMEANS
# ==========================================

kmeans = KMeans(
    n_clusters=2
)

kmeans.fit(X)

labels = kmeans.predict(X)

# ==========================================
# HIỂN THỊ KẾT QUẢ
# ==========================================

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    s=100
)

plt.xlabel("Income")

plt.ylabel("Spending")

plt.title("Phan nhom khach hang")

plt.show()
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# ==========================================
# DỮ LIỆU SINH VIÊN
# ==========================================

X = np.array([
    [8,9],
    [7,8],
    [2,3],
    [3,2]
])

# 1 = Dau
# 0 = Rot

y = np.array([
    1,
    1,
    0,
    0
])

# ==========================================
# HUẤN LUYỆN
# ==========================================

knn = KNeighborsClassifier(
    n_neighbors=3
)

knn.fit(X, y)

# ==========================================
# DỰ ĐOÁN
# ==========================================

new_student = np.array([
    [6,7]
])

result = knn.predict(
    new_student
)

print("Ket qua du doan:")

if result[0] == 1:
    print("Sinh vien DAU")
else:
    print("Sinh vien ROT")
```
