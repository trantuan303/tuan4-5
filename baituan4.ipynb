```python
# ==========================================
# DOC FILE MA TRAN KE
# ==========================================

def readGraph(filename):

    file = open(filename, "r")

    G = []

    for line in file:

        row = list(map(int, line.split()))

        G.append(row)

    file.close()

    return G

# ==========================================
# TINH BAC CAC DINH
# ==========================================

def degreeGraph(G):

    degree = []

    for i in range(len(G)):

        degree.append(sum(G[i]))

    return degree

# ==========================================
# SAP XEP DINH THEO BAC
# ==========================================

def sortNode(node, degree):

    sortedNode = []

    indeks = []

    for i in range(len(degree)):

        _max = 0

        for j in range(len(degree)):

            if j not in indeks:

                if degree[j] > _max:

                    _max = degree[j]

                    idx = j

        indeks.append(idx)

        sortedNode.append(node[idx])

    return sortedNode

# ==========================================
# TO MAU DO THI
# ==========================================

def graphColoring(G, node):

    t_ = {}

    for i in range(len(G)):

        t_[node[i]] = i

    degree = degreeGraph(G)

    sortedNode = sortNode(node, degree)

    colorDict = {}

    for i in range(len(G)):

        colorDict[node[i]] = [
            "Blue",
            "Red",
            "Yellow",
            "Green"
        ]

    theSolution = {}

    for n in sortedNode:

        setTheColor = colorDict[n]

        theSolution[n] = setTheColor[0]

        adjacentNode = G[t_[n]]

        for j in range(len(adjacentNode)):

            if adjacentNode[j] == 1:

                if setTheColor[0] in colorDict[node[j]]:

                    colorDict[node[j]].remove(
                        setTheColor[0]
                    )

    return theSolution

# ==========================================
# CHUONG TRINH CHINH
# ==========================================

G = readGraph("graph.txt")

node = "ABCDEF"

result = graphColoring(G, node)

print("KET QUA TO MAU:")

for t, w in sorted(result.items()):

    print("Dinh", t, "=", w)
```

```python
import itertools

# ==========================================
# MA TRAN CHI PHI
# ==========================================

graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

# ==========================================
# DANH SACH THANH PHO
# ==========================================

cities = [0, 1, 2, 3]

# ==========================================
# TIM CHU TRINH NGAN NHAT
# ==========================================

min_path = float('inf')

best_path = []

for path in itertools.permutations(cities):

    current_cost = 0

    for i in range(len(path) - 1):

        current_cost += graph[path[i]][path[i+1]]

    current_cost += graph[path[-1]][path[0]]

    if current_cost < min_path:

        min_path = current_cost

        best_path = path

# ==========================================
# IN KET QUA
# ==========================================

print("CHU TRINH TOT NHAT:")

print(best_path)

print("CHI PHI NHO NHAT:")

print(min_path)
```
